```python
import numpy as np
import cv2
```

```python
## VideoCapture(Number of web-cam used) 
## VideoCapture(Video file name) 
cap = cv2.VideoCapture(0)  

## frame -> image it-self(numpy array format)
## ret --> if the capture works properly/not\
while True:
    ret, frame = cap.read()
    cv2.imshow('frame', frame)
    
    ## cv2.waitKey(1) returns the ASCII character of the key pressed
    if cv2.waitKey(1) == ord('q'):  ## So if 'q' is pressed, then the video capture breaks
        break

cap.release() ## releasing the hold camera resource
cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture(0)  

while True:
    ret, frame = cap.read()
    width  = int(cap.get(3)) ## returns the float, so convert into int
    height = int(cap.get(3))
    
    image         = np.zeros(frame.shape, np.unit8)  ## an empty list with the captured frame rate
    smaller_frame = cv2.resize(frame, (0,0), fx=0.5, fy=0.5)

    ## top-left
    image[:heigth//2, :width//2] = smaller_frame
    
    ## bottom-left
    image[heigth//2, :width//2] = smaller_frame
    
    ## top-right
    image[:heigth//2, width//2:] = smaller_frame
    
    ## bottom-right
    image[heigth//2:, width//2:] = smaller_frame
    
    
    ## cv2.waitKey(1) returns the ASCII character of the key pressed
    if cv2.waitKey(1) == ord('q'):  ## So if 'q' is pressed, then the video capture breaks
        break

cap.release() ## releasing the hold camera resource
cv2.destroyAllWindows()
```
