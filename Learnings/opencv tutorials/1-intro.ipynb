## Re-sizing an image

```python
def display(img, heading):
    plt.title(heading)
    plt.imshow(img)
    plt.show()
```

```python
img = cv2.imread('assets/logo.jpg', 1)[:,:,::-1]  ## bgr to rgb
display(img, "NORMAL")
```

```python
img = cv2.resize(img, (0,0), fx=0.5, fy=0.5) ## Image is reduce to half of its size
display(img, "NORMAL")
```

## Rotate an image

```python
img1 = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)
display(img1, "ROTATE 90 CLOCKWISE")
```

```python
img3 = cv2.rotate(img, cv2.ROTATE_90_COUNTERCLOCKWISE)
display(img3, "ROTATE 90 COUNTERCLOCKWISE")
```

```python
img2 = cv2.rotate(img, cv2.ROTATE_180)
display(img2, "ROTATE 180")
```

## Saving an image

```python
cv2.imwrite('prashanth.jpg', img)
```
