```python
import cv2

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display(img):    
    plt.imshow(img)
    plt.show()
```

```python
## bgr representation --> blue, green, red
img1 = cv2.imread('assets/logo.jpg', 1)
img1 = cv2.imread('assets/logo.jpg', 1)[:,:,::-1]  ## bgr to rgb
```

```python
img1.shape

## 995 -->  number of rows     --> height of an image
## 1000 --> number of columns  --> width of an image
## 3 --->   number of channels --> color space of an image [How many values are representing a single pixel]
```

```python
## 3 values to represent a pixel
## regularly RGB, but in opencv it is BGR
```

```python
print(img1[0][0]) ## 1st pixel
print(img1[0][1]) ## 2nd pixel
```

## Changing the pixel value

```python
for i in range(100): ## row-wise (height of an image)
    for j in range(img1.shape[1]): ## column-wise (width of an image) --> here entire width is selected
        img1[i][j] = [ np.random.randint(0,255), np.random.randint(0,255), np.random.randint(0,255)]
display(img1)
```

## Copy 1 part of an image and Paste in another part of an image

```python
img1 = cv2.imread('assets/logo.jpg', 1)[:,:,::-1]  ## bgr to rgb
tag = img1[50:350, 400:700]
img1[300:600, 100:400] = tag

display(img1)
```
