```python
import cv2
import imutils
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python
def display(img):
    img =img[:,:,::-1]  ## rgb to bgr
    plt.imshow(img)
    plt.show()
```

```python
## Both are same images
img1 = cv2.imread("img1.jpg") 
img2 = cv2.imread("img1.jpg")
```

```python
display(img1)
```

```python
print("Shape : ",img1.shape)
print("Shape : ",img2.shape)
```

## Purposefully --> Applying some modifications in the image

```python
def adding_pixels(img):
    for _ in range(100):
        modified_height = np.random.randint(0,628)
        modified_width  = np.random.randint(0,498)
        img[modified_height][modified_width]= (0,255,0)

    return img
```

```python
img2 = adding_pixels(img2)
display(img2)
```

## Calculating the errors

```python
def mse(img1, img2):
    h, w, c = img1.shape
    diff    = cv2.subtract(img1, img2)
    err     = np.sum(diff**2)
    mse     = err/(float(h*w))
    return mse
```

```python
mse(img1, img2)
```

## Comparing pixels

```python
def compare_images(img1, img2):
    dict1 = {}
    for i in range(img1.shape[0]):
        for j in range(img1.shape[1]):
                if not np.array_equal(img1[i][j], img2[i][j]):
                    value = []
                    index = (i,j)
                    value.append(img1[i][j])
                    value.append(img2[i][j])
                    dict1[index] = value
    return dict1
```

```python
def extract_values(lst):
    return lst[0], lst[1]
```

```python
hash_dict = compare_images(img1, img2)
df = pd.DataFrame(list(hash_dict.items()), columns=['Index', 'image-1_and_image-2_pixel'])
df.head()
```

```python
df[['image-1 pixel', 'image-2 pixel']] = df['image-1_and_image-2_pixel'].apply(extract_values).apply(pd.Series)
df.drop('image-1_and_image-2_pixel', axis=1, inplace=True)

df.head()
```

## np.where() method

```python
diff = cv2.absdiff(img1, img2)
img2 = np.where(diff != 0, img1, img2)
```

```python
mse(img1, img2)
```
